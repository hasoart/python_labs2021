```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Упражнение 1
def evil(x):
    return np.log((np.exp(1/(np.sin(x)+1)))/(5/4+1/np.power(x, 15)))/np.log(1+x*x)

print(evil(np.array([1,10,1000])))
```

```python
# Упражнение 2
def f(x):
    return x*x - x - 6

x = np.linspace(-2.5, 4, 100)

fig, ax = plt.subplots()
ax.plot(x, f(x))
ax.grid(True, which='both')

ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
plt.show()
```

```python
# Упражнение 3
def even_more_evil(x):
    log_base = 1 + np.tan(1 / (1 + np.sin(x)**2))
    log_inside = (x**2 + 1) * np.exp(-0.1 * np.abs(x))
    return np.log(log_inside) / np.log(log_base)

x = np.linspace(-10, 10, 500)
plt.plot(x, even_more_evil(x))
plt.show()
```

```python
# Упражнение 4
f_string = input("Введите функцию: ")

def f(x):
    return eval(f_string)

X = np.linspace(-10, 10, 500)
Y = np.array([f(i) for i in X])
fig, ax = plt.subplots()
ax.plot(X, Y)
ax.grid(True, which='both')

ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
plt.show()
```

```python
# Упражнение 5
N = 20
X = np.arange(1, 1 + 20)
Y = (0.1*X*X) + np.random.normal(size=N)

plt.errorbar(X, Y, xerr=0.2, yerr=2)
a, b, c = np.polyfit(X, Y, deg=2)
plt.plot(X, a*X*X + b*X + c)
a, b = np.polyfit(X, Y, deg=1)
plt.plot(X, a*X + b)
plt.show()
```

```python
# Упражнение 6
def f(x, a=3, b=0.5, l=100):
    n = np.arange(0, l)
    return np.sum((b**n) * np.cos((a**n)*x*np.pi))

X = np.linspace(-1, 1, 1000)
plt.plot(X, np.array([f(i, l=1000) for i in X]))
plt.plot()
```

```python
def get_exit_coords(steps_num=10000, a=-5, b=5):
    steps = np.random.choice((-1, 1), steps_num)
    coords = np.cumsum(steps)

    a1 = np.argmax(coords>=b)
    a2 = np.argmax(coords<=a)

    if a1 == 0:
        exit_step = a2
        exit_side = 'a'
    elif a2 == 0:
        exit_step = a1
        exit_side = 'b'
    else:
        exit_step = min(a1, a2)
        exit_side = 'a' if a2 < a1 else 'b'
    
    return exit_step, exit_side
```

```python
# Мат. ожидание
np.mean([get_exit_coords(a=-5, b=5)[0] for i in range(10000)]) + 1
```

```python
a_count = 0
b_count = 0
for i in range(10000):
    side = get_exit_coords(a=-5, b=10)[1]
    if side == 'a':
        a_count += 1
    else:
        b_count += 1
print(f'Вероятность границы a :{a_count/10000}')
print(f'Вероятность границы b :{b_count/10000}')
```
